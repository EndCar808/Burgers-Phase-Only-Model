```python
import os
import numpy as np
import shutil
from subprocess import Popen, PIPE
from itertools import zip_longest
```

```python
os.getcwd()
cur_dir = "/work/projects/TurbPhase/burgers_1d_code/Burgers_PO"
```

```python
N = [64, 128, 256, 512];
alpha = np.arange(1.25, 2.5, 0.05);
k0 = [2];
beta = [0.0, 1.0];
u0 = "ALIGNED"
```

```python
i = 0
cmdList = []
for n in N:
    for k in k0:
        for b in beta:
            for a in alpha:
                input_dir_5 = "../Data/Snapshots/TriadDynamics/LCE_Runtime_Data_N[{}]_k0[{}]_ALPHA[{:0.3f}]_BETA[{:0.3f}]_u0[{}]_ITERS[400000]/SNAPS/Triad_Dynamics{}.mp4".format(n, k, a, b, u0, 5)
                input_dir_4 = "../Data/Snapshots/TriadDynamics/LCE_Runtime_Data_N[{}]_k0[{}]_ALPHA[{:0.3f}]_BETA[{:0.3f}]_u0[{}]_ITERS[400000]/SNAPS/Triad_Dynamics{}.mp4".format(n, k, a, b, u0, 4)
                output_dir = "../Data/Movies/Beta[{:0.3f}]_k0[{}]".format(b, k)
                
                filename = "/TriadDynamics_N[{}]_k0[{}]_ALPHA[{:0.3f}]_BETA[{:0.3f}]_u0[{}]_ITERS[400000].mp4".format(n, k, a, b, u0)
                
                if os.path.isfile(input_dir_5):
                    cmdList.append(["cp" + " " + input_dir_5 + " " + output_dir + filename])
                else:
                    cmdList.append(["cp" + " " + input_dir_4 + " " + output_dir + filename])        
                
                print(input_dir_5)
                print(output_dir)
                
                i += 1
```

```python
print(cmdList)
print(len(cmdList))
```

```python
procLimit = 10

# Create grouped iterable of subprocess calls to Popen() - see grouper recipe in itertools
groups = [(Popen(cmd, shell = True, stdout = PIPE, stdin = PIPE, universal_newlines = True) for cmd in cmdList)] * procLimit 

# # Loop of grouped iterable
for processes in zip_longest(*groups): 
    for proc in filter(None, processes): # filters out 'None' fill values if procLimit does not divide evenly into cmdList
        [runCodeOutput, runCodeErr] = proc.communicate()
        print(runCodeOutput)
        print(runCodeErr)
        proc.wait()
```
